# SELF ATTENTION
$\text{self-attention}=\text{softmax}(\frac{Q\cdot K^T}{\sqrt d_k} + M) \cdot V$

- $Q$: what I'm looking for?

- $K$: what I currently have?

- $V$: 

```python
def softmax(x):
    return np.exp(x)/ np.sum(np.exp(x), axis=1)
```

```python
def self_attention(sentence: np.ndarray):
    scaled = np.dot(sentence, sentence.T)/np.sqrt(len(sentence))
    m=np.tril(np.ones(scaled.shape))
    m[m==0]=-np.infty
    m[m==1]=0
    attention = softmax(scaled + m) # dot product similarity and softmax
    return attention
```

```python
attention_1= self_attention(em_sentence_1)
attention_2 = self_attention(em_sentence_2)
```
